```python

import random
opc = 0
montos = []
vehiculos = []
monto = round(random.uniform(1500, 3500), 2)
montos.append(monto)
monto = round(random.uniform(1500, 3500), 2)
montos.append(monto)
monto = round(random.uniform(1500, 3500), 2)
montos.append(monto)
def val_marca(marca):
  if len(marca) >= 2 and len(marca) <= 15:
    return True
def val_precio(precio):
  if precio > 5000000:
    return True
  else:
    return False
def grabar_vehiculo():
    patente = input("Ingrese la patente (Ejemplo: AB-CD-12):")
    tipo = input("Ingrese el tipo de vehículo:")
    marca = input("Ingrese la marca del vehículo:")
    precio = int(input("Ingrese el precio del vehículo:"))
    multas_monto = input("Ingrese el monto de las multas:")
    multas_fecha = input("Ingrese la fecha de las multas:")
    fecha_ingreso = input("Ingrese la fecha de ingreso:")
    nombre_dueno = input("Ingrese el nombre y apellido del dueño:")
    if not val_marca(marca):
     print("La marca ingresada no cumple con los requisitos. Intente nuevamente.")
     return
    if not val_precio(precio):
     print("El precio ingresado debe ser mayor a $5.000.000")
    vehiculo = {
    "patente": patente,
    "tipo": tipo,
    "marca": marca,
    "precio": precio,
    "multas": {
                "monto": multas_monto,
                "fechas": multas_fecha },
    "fecha_ingreso": fecha_ingreso,
    "nombre_dueño": nombre_dueno
    }
    vehiculos.append(vehiculo)
    print("El vehículo fue registrado")
def buscar_vehiculo():
    patente = input("Ingrese la patente del vehículo a buscar:")
    for vehiculo in vehiculos:
        if vehiculo["patente"] == patente:
            print("Información del vehículo:")
            for clave, valor in vehiculo.items():
                print(f"{clave.capitalize()}: {valor}")
            return
    print("No se encontro ningun vehículo con esa patente.")
def imprimir_certificados():
    patente = input("Ingrese la patente del vehículo:")
    for vehiculo in vehiculos:
        if vehiculo["patente"] == patente:
            certificados = ['Emisión de contaminantes', 'Anotaciones vigentes', 'Multas']
            for certificado, monto in zip(certificados, montos):
                print("Nombre del certificado:", certificado)
                print("Patente del vehículo:", patente)
                print("Datos del dueño actual:", vehiculo['nombre_dueño'])
                print("Monto:", monto)
                print()
            return
bienvenida = """Bienvenido a Auto Seguro"""
menu_opciones = """Indique la opción que desea:
1) Grabar datos del vehículo
2) Buscar vehículo por su patente
3) Imprimir certificados
4) Salir
"""
print(bienvenida)
while True:
  opc = input(menu_opciones)
  if opc == "1":
    grabar_vehiculo()
  elif opc == "2":
    buscar_vehiculo()
  elif opc == "3":
    imprimir_certificados()
  elif opc == "4":
   print("""
   Adios
   Programa creado por:
   Martin Zuazo
   Martin Osorio
   v.1.0.0
   """)
   break
```
